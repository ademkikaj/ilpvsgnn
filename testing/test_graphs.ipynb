```python
cora = pd.read_csv('docker/Benchmark/cora/relational/cora.csv')
content = pd.read_csv('docker/Benchmark/cora/relational/content.csv')
cites = pd.read_csv('docker/Benchmark/cora/relational/cites.csv')
```

```python
print("total number of examples: ", len(cora))
print(cora["class_label"].value_counts())
cora = cora[cora["class_label"].isin(["ML","Agents"])]
```

```python
# remove the paperid that are not in cora from content
content = content[content["paper_id"].isin(cora["paper_id"])]
# remover the citing_paper_id that are not in cora from cites
cites = cites[cites["citing_paper_id"].isin(cora["paper_id"])]
```

```python
# select 100 examples from each class
cora = cora.groupby('class_label', group_keys=False).apply(lambda x: x.sample(min(len(x), 100)))
# "Agens" class is renamed to "pos" and "ML" class is renamed to "neg"
cora["class_label"] = cora["class_label"].map({"ML": "neg", "Agents": "pos"})
# shuffle the data
cora = cora.sample(frac=1).reset_index(drop=True)
cora
```

```python
content.to_csv('docker/Benchmark/cora/relational/content.csv', index=False)
cites.to_csv('docker/Benchmark/cora/relational/cites.csv', index=False)
cora.to_csv('docker/Benchmark/cora/relational/cora.csv', index=False)
```
