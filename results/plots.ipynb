```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Test the scaling experiment
# Load the data
graph_df = pd.read_csv('bongard_gnn_scaling.csv')
logical_df = pd.read_csv('bongard_logic_scaling.csv')
print("Graph columns: ", graph_df.columns)
print("Logical columns: ", logical_df.columns)

merged_graph_df = graph_df.groupby(by='total_examples',as_index=False).max()

```

```python
plt.plot(merged_graph_df["total_examples"], merged_graph_df["runtime"], color='black')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Matplotlib Plot')
plt.grid(True, linestyle='--', color='gray', linewidth=0.5)
plt.show()
```

```python
plt.plot(logical_df["total_examples"], logical_df["time_total"], color='black')
plt.xlabel('X-axis')
```
