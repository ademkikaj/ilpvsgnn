```python
import os
```

```python
def display_file(dataset,representation,system):
    base_path = os.path.join("docker","Benchmark",dataset,representation,"logic",system)
    if system == "tilde":
        file_path = os.path.join(base_path,f"{dataset}.kb")
    elif system == "aleph":
        file_path = os.path.join(base_path,"background.pl")
    elif system == "popper":
        file_path = os.path.join(base_path,"bk.pl")
    
    # show the contents of the file
    with open(file_path) as f:
        print(f.read())
    return 

def show_bias_file(dataset,representation,system):
    base_path = os.path.join("docker","Benchmark",dataset,representation,"logic",system)
    file_path = os.path.join(base_path,"bias.pl")
    with open(file_path) as f:
        print(f.read())
    return

def show_hypothesis(dataset,representation,system):
    base_path = os.path.join("docker","Benchmark",dataset,representation,"logic",system)
    file_path = os.path.join(base_path,f"{dataset}_out.pl")
    with open(file_path) as f:
        print(f.read())
    return
```

```python
display_file("krk","Klog","tilde")
```

```python
dataset = "krk"
system = "aleph"
show_hypothesis(dataset,"node_edge",system)
show_hypothesis(dataset,"edge_based",system)
show_hypothesis(dataset,"Klog",system)
```
